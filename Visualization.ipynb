```python
import polars as pl
from polars import *
from matplotlib import pyplot as plt
```

# Fuzzy Set Visualisation

This notebook illustrates the membership function structures used in the project. Cells 1–3 provide brief reference examples of crisp vs. fuzzy membership and trapezoidal functions; Cell 4 shows the actual partition used in the demand forecasting model.

```python
demand_classical_set = pl.DataFrame(
    [
        (0,16),
        (0,17),
        (0,18),
        (0,19),
        (0,20),
        (1,21),
        (1,22),
        (1,23),
    ],
    ['high_demand', 'demand']
    ,orient='row'
)
```

## 1. Crisp Set (reference)

Membership is binary: elements below the threshold 21 have μ = 0, elements at or above have μ = 1.

```python
def plot_stepped(df,draw_style='steps-post',y='high_demand'):
    plot_df = df.to_pandas()
    plot_df.plot(kind='line', x='demand', y=y, linewidth=5, xlabel='Demand', ylabel='High demand',
                 drawstyle=draw_style)
    plt.show()


plot_stepped(demand_classical_set)
```

```python
demand_fuzzy_set = pl.DataFrame(
    [
        (0,16),
        (0.2,17),
        (0.4,18),
        (0.6,19),
        (0.8,20),
        (1,21),
        (1,22),
        (1,23),
    ],
    ['high_demand', 'demand']
    ,orient='row'
)
```

## 2. Fuzzy Set (reference)

Membership grades are continuous on [0, 1], allowing partial membership. Here the "high demand" set rises linearly from μ = 0 at 16 to μ = 1 at 21.

```python
plot_stepped(demand_fuzzy_set,'default')
```

```python
trapezoid_function = pl.DataFrame(
    [
        (0,0),
        (2,0),
        (4,0.2),
        (6,0.4),
        (8,0.6),
        (10,0.8),
        (12,1),
        (14,0.8),
        (16,0.6),
        (18,0.4),
        (20,0.2),
        (22,0),
        (24,0),
    ],
    ['demand', 'mediocre_demand']
    ,orient='row'
)
```

## 3. Trapezoidal Membership Function (reference)

A trapezoidal function has a flat top (μ = 1 over an interval) and linear slopes on both sides. A triangular function is the degenerate case where the flat top collapses to a single point.

```python
plot_stepped(trapezoid_function,'default','mediocre_demand')
```

```python
fuzzy_sets = pl.DataFrame(
      [
          (0,   1, 0, 0, 0, 0),
          (25,  0, 1, 0, 0, 0),
          (50,  0, 0, 1, 0, 0),
          (75,  0, 0, 0, 1, 0),  # added 0
          (100, 0, 0, 0, 0, 1),  # added 1
      ],
      ['demand', 'very_low_demand', 'low_demand', 'medium_demand', 'high_demand', 'very_high_demand'],
      orient='row'
  )
```

## 4. Demand Partition Used in This Project

The model partitions the demand universe into five fuzzy sets. Boundary sets (VeryLow, VeryHigh) use shoulder functions so that the most extreme demand values always receive full membership — preventing boundary regions from being systematically under-fired during inference.

```python
fuzzy_sets.to_pandas().plot(kind='line',x='demand',linewidth=5,xlabel='Demand',ylabel='Membership')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```
